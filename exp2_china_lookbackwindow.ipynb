```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10, 6))

df = pd.read_csv('exp1_results/lookback/china.csv')
# df.plot(x='Input-size',y='Autoformer',marker='o', ax=ax)
markers = ['o', 's', 'D', 'v', '^', 'p', '*']  # 異なるマーカーの種類
colors = ['r', 'b', '#999933', '#03AF7A', '#990099', '#F6AA00', 'k']  # 異なる色の種類
# colors = ['#FF0000', '#00FF00', '#0000FF', '#00FFFF', '#FF00FF', '#FFFF00', '#000000']  # 異なる色の種類

for i, column in enumerate(df.columns[1:]):
    df.plot(x='Input-size', y=column,linestyle='--',marker=markers[i], color=colors[i],ax=ax,markersize=8,linewidth=2.3)

ax.legend(df.columns[1:], loc='lower center', bbox_to_anchor=(0.5, 1), frameon=False, fancybox=False, shadow=False, ncol=4)
ax.grid(axis='x', linestyle='--')
ax.legend([ 'Autoformer','Informer','Transformer',  'DLinear',  'Non-stationary Transformer','FEDformer', 'Linear'], loc='lower center', bbox_to_anchor=(0.5, 1), frameon=False, fancybox=False, shadow=False, ncol=4,fontsize=14)

ax.grid(axis='x', linestyle='--')
ax.set_xlabel('Look-back window size L',fontsize=18)
ax.set_ylabel('MSE',fontsize=18)
ax.set_ylim(0.5,1.15)
plt.savefig('exp1_results/lookback/china.png')
```

```python
df.drop('DLinear', axis=1, inplace=True)

fig, ax = plt.subplots(figsize=(10, 6))
markers = ['o', 's', 'D', 'v', '^', 'p']  # 異なるマーカーの種類
colors = ['b', '#03AF7A','r','#990099',   '#F6AA00','#999933',]  # 異なる色の種類
# colors = ['#FF0000', '#00FF00', '#0000FF', '#00FFFF', '#FF00FF', '#FFFF00', '#000000']  # 異なる色の種類

for i, column in enumerate(df.columns[1:]):
    df.plot(x='Input-size', y=column,linestyle='--',marker=markers[i], color=colors[i],ax=ax,markersize=8,linewidth=2.3)

ax.legend(df.columns[1:], loc='lower center', bbox_to_anchor=(0.5, 1), frameon=False, fancybox=False, shadow=False, ncol=4)
ax.grid(axis='x', linestyle='--')
ax.legend([ 'Autoformer','Informer','Transformer','Non-stationary Transformer','FEDformer', 'Linear'], loc='lower center', bbox_to_anchor=(0.5, 1), frameon=False, fancybox=False, shadow=False, ncol=6,fontsize=20)

ax.grid(axis='x', linestyle='--')
ax.set_xlabel('Look-back window size L',fontsize=18)
ax.set_ylabel('MSE',fontsize=18)
plt.savefig('exp1_results/lookback/china_excDL.png',bbox_inches='tight')
plt.show()
```
